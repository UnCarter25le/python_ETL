```python
#教育部租屋--https://csrc.edu.tw/NewCsrcHouse/RentHouse/RentQA?page=1
#抓網址

from bs4 import BeautifulSoup
from splinter import Browser
import requests
import splinter
import json
import time

url = 'https://csrc.edu.tw/NewCsrcHouse/RentHouse/RentQA?page=1'
# executable_path = {'executable_path':'/usr/local/bin/phantomjs'}
# browser = Browser('phantomjs', **executable_path)
browser = splinter.Browser(driver_name='chrome')   #選擇以google chrome 而不是phantomjs
browser.visit(url)

# browser.html
toool = []
for i in range(1,4):
    try:
        time.sleep(1)
        for w in range(1,11): 
            element = browser.find_by_xpath('//*[@id="fragment-2"]/table/tbody/tr[1]/td/table/tbody/tr['+str(w)+']/td[2]/a')
            toool.append(element['href'])
        browser.find_by_xpath('//*[@id="fragment-2"]/table/tbody/tr[5]/td/div/a[3]').click()
    except:
        print('over')
i += 1
print(len(toool))
```

```python
#抓每個網址的網頁

from bs4 import BeautifulSoup
from splinter import Browser
import requests
import splinter
import json
import time

browser = splinter.Browser(driver_name='chrome')

t1 = time.time()

jsonarray = []
for i in range(len(toool)):
    time.sleep(1)
    jsondict = {}
    jsondict['url'] = toool[i]
    
    url = toool[i]
    browser.visit(url)
    jsondict['html'] = browser.html   #以splinter擷取整個畫面
    browser.find_by_css('#mt10 > table:nth-child(3) > tbody > tr > td > table > tbody > tr:nth-child(2) > td > table > tbody \
    > tr:nth-child(5) > td > table > tbody > tr > td:nth-child(5) > a > font:nth-child(2) > b').mouse_out()
    
    jsondict['update'] = '2017-08-19'
    jsonarray.append(jsondict)
browser.quit()
    
    print(len(jsonarray))
    
with open('educationparty_24.json','w') as f:
    json.dump(jsonarray,f)
    
t2 = time.time()
print(t2-t1)
```
