```python
#抓台北每個網址的網頁

from bs4 import BeautifulSoup
from splinter import Browser
import requests
import splinter
import json
import time


with open('./textmining/租屋照妖鏡_taipei_url.json') as D:
    json_in = D.read()
    json_out = json.loads(json_in)

t1 = time.time()

jsonarray = []

for i in range(len(json_out)):
    url = json_out[i]
    res = requests.get(url)
    res.encoding = 'utf-8'
    
    executable_path = {'executable_path':'/usr/local/bin/phantomjs'}
    browser = Browser('phantomjs', **executable_path)
#     browser = splinter.Browser(driver_name='chrome')
    time.sleep(1)
    jsondict = {}
    jsondict['url'] = json_out[i]
    
    browser.visit(url)
    jsondict['html'] = browser.html
    
    soup = BeautifulSoup(res.text,'lxml').select('time')[0].text.replace('\n','').replace(' ','').split(',')
    jsondict['update'] = soup[1]+'-0'+soup[0].split('月')[0]+'-'+soup[0].split('月')[1]
    
    jsonarray.append(jsondict)
    time.sleep(1)
#     browser.quit()
    
    print(len(jsonarray))

with open('./textmining/租屋照妖鏡_taipei.json','w') as f:
    json.dump(jsonarray,f)
    
t2 = time.time()
print(t2-t1)
D.close()
```
