# 製作django用的sqlite3


```python
import sqlite3
```

```python
#建立資料庫連線。如果沒有找到對應的資料庫，就會幫忙創建一個。檔名不一定要是db.sqlite3，可以是DB.sqlite3
db = sqlite3.connect('db.sqlite3')
print(db)
```

```python
#建立寫入手
cur = db.cursor()
```

```python
#藉由寫入手執行，創立資料表table  ,  table name daanmarkets
#auto produce exclusive id
#please abide by the pattern: cur.execute(  "sql;"  )  
cur.execute("CREATE TABLE daanmarkets(id INTEGER PRIMARY KEY AUTOINCREMENT,\   
            mapName varchar,\
            mapLat varchar,\
            mapLng varchar,\
            mapAddress varchar,\
            mapcityID varchar);")
```

```python
#table name daanfiredepart

cur.execute("CREATE TABLE daanfiredepart(id INTEGER PRIMARY KEY AUTOINCREMENT,\
            mapName varchar,\
            mapLat varchar,\
            mapLng varchar,\
            mapAddress varchar,\
            mapcityID varchar);")
```

```python
#table name daannoise

cur.execute("CREATE TABLE daannoise(id INTEGER PRIMARY KEY AUTOINCREMENT,\
            mapName varchar,\
            mapLat varchar,\
            mapLng varchar,\
            mapAddress varchar,\
            mapcityID varchar);")
```

# 測試插入多筆
