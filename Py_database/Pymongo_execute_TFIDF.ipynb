# QA內容整理

```python
#針對qa的content

from collections import Counter
import nltk
import jieba
from jieba.analyse import extract_tags as et
import jieba.posseg as pseg
import codecs

jieba.load_userdict('./dict/jieba_tranditional_dict.txt')
jieba.load_userdict('./dict/QA.txt')
stopkey=[line.strip() for line in open('./stopwords.txt').readlines()]   #set stopwords
enforce =[line.strip() for line in open('./dict/QA.txt').readlines()]   #enforce specific words weights

wd_count = Counter() 
QA_array = []
for row in collection2.find({},{'content':1,'_id':0}):
#有設停用字
    wordclean = ''
    for word in jieba.cut(row['content'],cut_all=False):
        if word not in stopkey:
            word.replace(' ','')
            wordclean += word + ' '
            if word in enforce:         #=========enforce portion========
                if word in wd_count:
                    wd_count[word] += 1
                else:
                    wd_count[word] = 1*200
            if word not in enforce:     #=======not in enforce portion======
                if word in wd_count:
                    wd_count[word] += 1
                else:
                    wd_count[word] = 1 
    QA_array.append(wordclean)
    
# tags = et(corpus, topK=10)  
# print(u"提取关键词結果:")  
# print(" ".join(tags))

wd_count.most_common(200)
```

# TFIDF

```python
import time
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import CountVectorizer  

vectorizer = CountVectorizer()
transformer = TfidfTransformer()
tfidf = transformer.fit_transform(vectorizer.fit_transform(QA_array))
                 #TF-IDF               #TF
print(tfidf.shape)

words = vectorizer.get_feature_names()

# consequent
start = time.time()
for i in range(len(QA_array)):
    print('----Document %d----' % (i))
    for j in range(len(words)):
        if tfidf[i, j] > 0.3:
            print(words[j], tfidf[i, j])
end = time.time()
print(int(end-start))


#if u wanna calculate cosine_similarity, u can use package as bellow:
# from sklearn.metrics.pairwise import cosine_similarity
# cosine_similarity(tfidf[0:1], tfidf)
```
