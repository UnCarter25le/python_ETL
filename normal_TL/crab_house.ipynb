# 抓蟹居網物件網址 https://rent.tmm.org.tw/

```python
#台北市分租套房

from bs4 import BeautifulSoup
from splinter import Browser
import requests
import splinter
import json
import time

t1 = time.time()  #紀錄時間
final_1 = []

URL = 'https://rent.tmm.org.tw/'
executable_path = {'executable_path':'/usr/local/bin/phantomjs'}  #利用phantomjs的必要路徑
browser = Browser('phantomjs', **executable_path)       #瀏覽器換成phantomjs

# browser = splinter.Browser(driver_name='chrome')      #如果要看爬取的狀況，可將瀏覽器設成google chrome
browser.encoding = 'utf-8'

# 設定splinter點擊方法
def brow(a,b,c,d):
    browser.visit(a)
    browser.find_option_by_text(b).first.click()
    browser.find_option_by_text(c).first.click()
    browser.find_by_name(d).click()
    print('open_already')

brow(URL,'臺北市','套房','submit')
w = 1
while w < 4:
    for i in range(2,40):
        try:              #利用xpath來找到hyperlink，以key-value的方式抓取下來（element_1['herf']）
            element_1 = browser.find_by_xpath('//*[@id="divTagResult"]/div/ul['+str(i)+']/li[1]/div/div[3]/a')
            final_1.append(element_1['href'])  
        except:
            if w == 3:     
                print('over1')
                break          #escape for loop
            else:
                browser.find_link_by_text('Next').click()  #跳下一頁
                print('go_to')
                break          #escape for loop
    w += 1
print('累積'+str(len(final_1)))


brow(URL,'臺北市','獨立套房','submit')
w = 1
while w < 4:
    for i in range(2,40):
        try:
            element_1 = browser.find_by_xpath('//*[@id="divTagResult"]/div/ul['+str(i)+']/li[1]/div/div[3]/a')
            final_1.append(element_1['href'])  
        except:
            if w == 3:     
                print('over2')
                break
            else:
                browser.find_link_by_text('Next').click()
                print('go_to')
                break
    w += 1
print('累積'+str(len(final_1)))

brow(URL,'臺北市','雅房','submit')
w = 1
while w < 6:
    for i in range(2,40):
        try:
            element_1 = browser.find_by_xpath('//*[@id="divTagResult"]/div/ul['+str(i)+']/li[1]/div/div[3]/a')
            final_1.append(element_1['href'])  
        except:
            if w == 5:     
                print('over3')
                break
            else:
                browser.find_link_by_text('Next').click()
                print('go_to')
                break
    w += 1
print('累積'+str(len(final_1)))

brow(URL,'臺北市','住家','submit')
w = 1
while w < 6:
    for i in range(2,40):
        try:
            element_1 = browser.find_by_xpath('//*[@id="divTagResult"]/div/ul['+str(i)+']/li[1]/div/div[3]/a')
            final_1.append(element_1['href'])  
        except:
            if w == 5:     
                print('over4')
                break
            else:
                browser.find_link_by_text('Next').click()
                print('go_to')
                break
    w += 1
print('累積'+str(len(final_1)))

#新北市
brow(URL,'新北市','套房','submit')
w = 1
while w < 3:
    for i in range(2,40):
        try:
            element_1 = browser.find_by_xpath('//*[@id="divTagResult"]/div/ul['+str(i)+']/li[1]/div/div[3]/a')
            final_1.append(element_1['href'])  
        except:
            if w == 2:     
                print('over5')
                break
            else:
                browser.find_link_by_text('Next').click()
                print('go_to')
                break
    w += 1
print('累積'+str(len(final_1)))


brow(URL,'新北市','獨立套房','submit')
w = 1
while w < 2:
    for i in range(2,40):
        try:
            element_1 = browser.find_by_xpath('//*[@id="divTagResult"]/div/ul['+str(i)+']/li[1]/div/div[3]/a')
            final_1.append(element_1['href'])  
        except:
            if w == 1:     
                print('over6')
                break
            else:
                browser.find_link_by_text('Next').click()
                print('go_to')
                break
    w += 1
print('累積'+str(len(final_1)))

brow(URL,'新北市','雅房','submit')
w = 1
while w < 3:
    for i in range(2,40):
        try:
            element_1 = browser.find_by_xpath('//*[@id="divTagResult"]/div/ul['+str(i)+']/li[1]/div/div[3]/a')
            final_1.append(element_1['href'])  
        except:
            if w == 2:     
                print('over7')
                break
            else:
                browser.find_link_by_text('Next').click()
                print('go_to')
                break
    w += 1
print('累積'+str(len(final_1)))

brow(URL,'新北市','住家','submit')
w = 1
while w < 5:
    for i in range(2,40):
        try:
            element_1 = browser.find_by_xpath('//*[@id="divTagResult"]/div/ul['+str(i)+']/li[1]/div/div[3]/a')
            final_1.append(element_1['href'])  
        except:
            if w == 4:     
                print('over8')
                break
            else:
                browser.find_link_by_text('Next').click()
                print('go_to')
                break
    w += 1
print('累積'+str(len(final_1)))



t2 = time.time()
print(t2-t1)

browser.quit()
```
