```python
#洗成總攬面貌
import json
import requests
from bs4 import BeautifulSoup

with open('./textmining/QA/崔媽媽_QA.json') as f:
    inn = f.read()
    outt = json.loads(inn)

soup = BeautifulSoup(outt[0]['html'],'lxml')
soup1 = soup.select('h2')[1:]#6分類
soup2 = soup.select('h3')#6 每類的細述（扣掉前言）

#回答內容的id跟分類與問題數有關  1sub4代表第一類的第四題
# 1,4  第一類有4個回答
# 2,13
# 3,5
# 4,2
# 5,7
# 6,3
soup3 = soup.select_one('#1sub4').text
soup4 = soup.select('font > span') #34個問題
soup5 = []  #將問題洗好
for i in range(34):
    soup5.append(soup4[i].text.replace('\r\n\t\t\t','').replace(' ','').replace('\r\n','').\
                 replace('\t\t','')[2:].replace('.','').replace('0','').replace('1','').replace('2','').\
                replace('3',''))
#將問題依照分類兒分類
soup5_first = soup5[0:4]
soup5_second = soup5[4:17]
soup5_third = soup5[17:22]
soup5_fourth = soup5[22:24]
soup5_fifth = soup5[24:31]
soup5_sixth = soup5[31:34]
jsondict = {}

jsondict['url'] = outt[0]['url']
jsondict['update'] = outt[0]['update']

for i in range(4):
    w = 1
    w += i
    jsondict[soup5_first[i]] = soup.select_one('#1sub'+str(w)).text.replace('\u3000','').replace(' ','').replace('\r\n','').replace('\t\t','').replace('\n','').replace('\n\n\n','')


for i in range(13):
    w = 1
    w += i
    jsondict[soup5_second[i]] = soup.select_one('#2sub'+str(w)).text.replace('\u3000','').replace(' ','').replace('\r\n','').replace('\t\t','').replace('\n','').replace('\n\n\n','')


for i in range(5):
    w = 1
    w += i
    jsondict[soup5_third[i]] = soup.select_one('#3sub'+str(w)).text.replace('\u3000','').replace(' ','').replace('\r\n','').replace('\t\t','').replace('\n','').replace('\n\n\n','')


for i in range(2):
    w = 1
    w += i
    jsondict[soup5_fourth[i]] = soup.select_one('#4sub'+str(w)).text.replace('\u3000','').replace(' ','').replace('\r\n','').replace('\t\t','').replace('\n','').replace('\n\n\n','')


for i in range(7):
    w = 1
    w += i
    jsondict[soup5_fifth[i]] = soup.select_one('#5sub'+str(w)).text.replace('\u3000','').replace(' ','').replace('\r\n','').replace('\t\t','').replace('\n','').replace('\n\n\n','')


for i in range(3):
    w = 1
    w += i
    jsondict[soup5_sixth[i]] = soup.select_one('#6sub'+str(w)).text.replace('\u3000','').replace(' ','').replace('\r\n','').replace('\t\t','').replace('\n','').replace('\n\n\n','')


with open('崔媽媽QA.json','w') as f:
    json.dump(jsondict,f)
```

```python
#請先執行完上面的程式
# 洗成complete
import json
with open('./textmining/QA/崔媽媽QA.json') as f:
    inn = f.read()
    outt = json.loads(inn)
    
jsonarray = []

key = list(outt.keys())[2:]
value = list(outt.values())[2:]

for i in range(0,4):
    jsondict = {}
    jsondict['url'] = outt['url']
    jsondict['update'] = outt['update']
    jsondict['name'] = key[i]
    jsondict['content'] = value[i]
    jsondict['genre'] = soup1[0].text.replace('【','').replace('】','')
    jsondict['genre_description'] = soup2[0].text.replace(' ','')
    jsonarray.append(jsondict)
for i in range(4,17):
    jsondict = {}
    jsondict['url'] = outt['url']
    jsondict['update'] = outt['update']
    jsondict['name'] = key[i]
    jsondict['content'] = value[i]
    jsondict['genre'] = soup1[1].text.replace('【','').replace('】','')
    jsondict['genre_description'] = soup2[1].text.replace(' ','')
    jsonarray.append(jsondict)
for i in range(17,22):
    jsondict = {}
    jsondict['url'] = outt['url']
    jsondict['update'] = outt['update']
    jsondict['name'] = key[i]
    jsondict['content'] = value[i]
    jsondict['genre'] = soup1[2].text.replace('【','').replace('】','')
    jsondict['genre_description'] = soup2[2].text.replace(' ','')
    jsonarray.append(jsondict)
for i in range(22,24):
    jsondict = {}
    jsondict['url'] = outt['url']
    jsondict['update'] = outt['update']
    jsondict['name'] = key[i]
    jsondict['content'] = value[i]
    jsondict['genre'] = soup1[3].text.replace('【','').replace('】','')
    jsondict['genre_description'] = soup2[3].text.replace(' ','')
    jsonarray.append(jsondict)
for i in range(24,31):
    jsondict = {}
    jsondict['url'] = outt['url']
    jsondict['update'] = outt['update']
    jsondict['name'] = key[i]
    jsondict['content'] = value[i]
    jsondict['genre'] = soup1[4].text.replace('【','').replace('】','')
    jsondict['genre_description'] = soup2[4].text.replace(' ','')
    jsonarray.append(jsondict)
for i in range(31,34):
    jsondict = {}
    jsondict['url'] = outt['url']
    jsondict['update'] = outt['update']
    jsondict['name'] = key[i]
    jsondict['content'] = value[i]
    jsondict['genre'] = soup1[5].text.replace('【','').replace('】','')
    jsondict['genre_description'] = soup2[5].text.replace(' ','')
    jsonarray.append(jsondict)
    
with open('崔媽媽QA_complete.json','w') as G:
    json.dump(jsonarray,G)
```
