```python
#splinter -newtaipei city


from bs4 import BeautifulSoup
import requests
import splinter
import json
import time

tool1 = []        #製作jsonarray用

t1 = time.time()  #紀錄時間

#（一）抓取各學校代號清單
url = 'http://house.nfu.edu.tw/protal/all/3.html'
res = requests.get(url)
res.encoding = 'utf-8'
soup = BeautifulSoup(res.text,'lxml').select('.readmore > a') 
schoollist = [row.text[2:] for row in soup]

#（二）進到各學校，抓取各物件的詳細資訊連結
for w in range(len(schoollist)):
    URL = 'http://house.nfu.edu.tw/house/{}.html'.format(schoollist[w])
    browser = splinter.Browser(driver_name='chrome')   #以splinter模擬
    browser.visit(URL)
    time.sleep(1)
    
    try: #寫try except可以讓程式碰到索引超過的時候，依然進行
        for i in range(1,2000):
            browser.click_link_by_id('gmapSidebarItem_'+str(i))       #點擊右側欄位
            browser.find_by_id('gmapSidebarItem_'+str(i)).mouse_out() #模擬滑鼠移開
            time.sleep(1)
            element = browser.find_link_by_partial_text('詳細資訊')    #抓取連結
            tool1.append(element['href'])                           
            browser.find_link_by_partial_text('詳細資訊').mouse_out()  #模擬滑鼠移開

    except Exception as e:
        print(schoollist[w]+'  '+'out of range or no Object')

    print(schoollist[w]+'  done'+'  目前共'+str(len(tool1)))           #目前總共抓了幾個網址
    
    browser.quit()                      #抓完一個學校的物件，瀏覽器要關閉
    
    
t2 = time.time()     #計算消耗時間
print(t2-t1)



#（三）利用phantomjs讓splinter背景執行---進到各物件的連結，抓取租屋的資訊
from splinter import Browser
executable_path = {'executable_path':'/usr/local/bin/phantomjs'}  #要先安裝好phantomjs。這兩行是基本的程式設定。
browser = Browser('phantomjs', **executable_path)

tool2 = []   #製作jsonarray用

try:
    for link in tool1:  #tool1裝的是所有詳細資訊的連結        
        
        cloudHTML = {}  #製作json用

        time.sleep(1)

        res = requests.get(link)
        res.encoding = 'utf-8'
        soup1 = BeautifulSoup(res.text,'lxml').select_one('#info') 
        
        #存網頁連結，物件更新時間（年月日），網頁內容
        cloudHTML['url'] = link
        cloudHTML['update'] = soup1.text.split('\n')[2].split('／')[0][8:19] 
        cloudHTML['html'] = res.text
        
        tool2.append(cloudHTML)   #將一個個租屋物件存進array
        
        print(len(tool2))   #紀錄完成幾筆
except Exception as e:
     print('out of range')

with open('cloudrent_newtaipei.json','w') as f:   #寫出檔案
    json.dump(tool2,f)

t3 = time.time()  #計算消耗時間
print(t3-t1)
```

```python
#google api 轉經緯度，郵遞區號

import requests


api_url= "https://maps.googleapis.com/maps/api/geocode/json?address={}&key="
def get_location(address):
    URL=api_url.format(address)
    res= requests.get(URL)
    location_dict = res.json()
    lat = location_dict['results'][0]['geometry']['location']['lat']
    lng = location_dict['results'][0]['geometry']['location']['lng']
    cityID = location_dict['results'][0]['address_components'][-1]['short_name']
    res.close()
    return lat,lng,cityID

address = '台北市信義區永吉路30巷'
get_location(address)  #tuple
```
